#Text Pre-Processing

```python
import spacy
import nltk
```

```python
import spacy.cli
spacy.cli.download("en_core_web_lg")
nlp = spacy.load("en_core_web_lg")
```

```python
# 'nlp_docs' will be a list of spaCy Doc objects, each corresponding to a processed review
nlp_docs = [nlp(text) for text in filtered_df['text']]
```

```python
# 'sentence_list_spacy' will be a list of sentences across all the reviews
sentence_list_spacy = [sentence for doc in nlp_docs for sentence in doc.sents]
```

```python
# 'token_list_spacy' will be a list of all tokens (words and punctuation) from all reviews
token_list_spacy = [token for doc in nlp_docs for token in doc]
```

```python

#finding top 20 nouns, verbs, adjective and named entity recognition
```

```python
#Fetching 1 star and 5 star reviews

one_star_reviews = filtered_df[filtered_df['stars'] == 1]['text']
five_star_reviews = filtered_df[filtered_df['stars'] == 5]['text']

```

```python
#Finding the top 20 frequently used nouns, adjectives and verbs in 1 star reviews and 5 star reviews, respectively

def process_reviews(reviews):
    adj_freq = Counter()
    verb_freq = Counter()
    noun_freq = Counter()

    for review in reviews:
        doc = nlp(review)
        for token in doc:
            if token.pos_ == 'ADJ':
                adj_freq[token.lemma_.lower()] += 1
            elif token.pos_ == 'VERB':
                verb_freq[token.lemma_.lower()] += 1
            elif token.pos_ == 'NOUN':
                noun_freq[token.lemma_.lower()] += 1

    return adj_freq, verb_freq, noun_freq
```

```python
one_star_reviews = filtered_df[filtered_df['stars'] == 1]['text']
five_star_reviews = filtered_df[filtered_df['stars'] == 5]['text']

adj_freq_1_star, verb_freq_1_star, noun_freq_1_star = process_reviews(one_star_reviews)
adj_freq_5_star, verb_freq_5_star, noun_freq_5_star = process_reviews(five_star_reviews)
```

```python
top_20_adjectives_1_star = adj_freq_1_star.most_common(20)
top_20_verbs_1_star = verb_freq_1_star.most_common(20)
top_20_nouns_1_star = noun_freq_1_star.most_common(20)

top_20_adjectives_5_star = adj_freq_5_star.most_common(20)
top_20_verbs_5_star = verb_freq_5_star.most_common(20)
top_20_nouns_5_star = noun_freq_5_star.most_common(20)
```

```python
print("Top 20 Adjectives in 1-Star Reviews:", top_20_adjectives_1_star)
print("Top 20 Verbs in 1-Star Reviews:", top_20_verbs_1_star)
print("Top 20 Nouns in 1-Star Reviews:", top_20_nouns_1_star)

print("Top 20 Adjectives in 5-Star Reviews:", top_20_adjectives_5_star)
print("Top 20 Verbs in 5-Star Reviews:", top_20_verbs_5_star)
print("Top 20 Nouns in 5-Star Reviews:", top_20_nouns_5_star)
```

#Finding the top 20 frequently used named entities from the selected reviews.

```python
def process_reviews_for_entities(reviews):
    entity_freq = Counter()

    for review in reviews:
        doc = nlp(review)
        for ent in doc.ents:
            entity_freq[ent.text.lower()] += 1

    return entity_freq
```

```python
# For the entire dataset
all_reviews = filtered_df['text']
all_entities_freq = process_reviews_for_entities(all_reviews)
top_20_entities = all_entities_freq.most_common(20)

# Or separately for 1-star and 5-star reviews
one_star_reviews = filtered_df[filtered_df['stars'] == 1]['text']
five_star_reviews = filtered_df[filtered_df['stars'] == 5]['text']

entities_freq_1_star = process_reviews_for_entities(one_star_reviews)
entities_freq_5_star = process_reviews_for_entities(five_star_reviews)

top_20_entities_1_star = entities_freq_1_star.most_common(20)
top_20_entities_5_star = entities_freq_5_star.most_common(20)
```

```python
# For the entire dataset
print("Top 20 Named Entities:", top_20_entities)

# Or separately for 1-star and 5-star reviews
print("Top 20 Named Entities in 1-Star Reviews:", top_20_entities_1_star)
print("Top 20 Named Entities in 5-Star Reviews:", top_20_entities_5_star)
```

#Summary

The analysis of top words in 1-star and 5-star reviews reveals distinct patterns: 1-star reviews predominantly feature negative adjectives ('bad', 'rude', 'terrible') and nouns pointing to service issues ('order', 'time', 'minute'), whereas 5-star reviews are characterized by positive descriptors ('good', 'great', 'delicious') and nouns emphasizing overall experience ('food', 'place', 'service'). This linguistic contrast highlights differing customer experiences and expectations. Notably, named entities like 'Tucson' and various cuisines ('Italian', 'Chinese', 'Thai') appear across both, suggesting location and food type as common discussion points, but with starkly different sentiments in each review category.
